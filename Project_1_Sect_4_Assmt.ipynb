Prepare

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/mtsilimos/url/main/houseprices.csv"
dataset = pd.read_csv(url)
dataset
```

Drop columns with less than 700 values

```python
dataset.dropna(axis=1, thresh = 700, inplace = True)
```

Perform imputation of median values for missing numeric values:

```python
dataset = dataset.fillna(dataset.median(numeric_only=True))
```

Identify Numerical Columns

```python
numerical_columns = dataset.select_dtypes(include='number')
print(numerical_columns.columns.tolist())
```

Analyze numerical columns

Include more numeric variables from the dataset ‘House Prices’ in your model.

```python
df = dataset[['LotFrontage', 'LotArea', 'OverallQual', 'OverallCond', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'TotalBsmtSF', '1stFlrSF', 'GrLivArea', 'BedroomAbvGr', 'TotRmsAbvGrd', 'SalePrice']]
```

Select features to be used for regression

```python
X = dataset[['LotFrontage', 'LotArea', 'OverallQual', 'OverallCond', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'TotalBsmtSF', '1stFlrSF', 'GrLivArea', 'BedroomAbvGr', 'TotRmsAbvGrd']]
y = dataset['SalePrice']           
```

Split dataset into training and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

Scale data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().set_output(transform="pandas")
```

```python
X_train = scaler.fit_transform(X_train, y_train)
```

```python
X_test = scaler.transform(X_test)
```

Perform Ridge Regularisation because I suspect that many of the features might be correlated, but I would not want to eliminate them completely from the model

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=20)
ridge.fit(X_train, y_train)
pd.Series(ridge.coef_, index = X_train.columns)
```

Perform cross validation to find optimum alpha for ridge regression

```python
from sklearn.linear_model import RidgeCV
ridge_cv = RidgeCV(alphas=[0.1, 0.001, 0.0001, 0.005, 1, 0.3, 0.5, 0.8, 10, 20, 100, 500])
ridge_cv.fit(X_train, y_train)

ridge_cv.alpha_
```

Train Rigde regression model with alpha = 100 and calculate R2 score

```python
from sklearn.metrics import r2_score
ridge = Ridge(alpha=100)
ridge.fit(X_train, y_train)
y_pred= ridge.predict(X_test)
r2 = r2_score(y_test, y_pred)
r2
```
